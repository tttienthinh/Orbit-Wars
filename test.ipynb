```python
N, Q = input().split()
N = int(N)
Q = int(Q)

liste = [chr(65 + i) for i in range(N)]


def merge(L1, L2, q):
    n = len(L1)
    m = len(L2)
    i = 0
    j = 0
    L = []
    while i<n and j<m and q>0:
        val1 = L1[i]
        val2 = L2[j]
        val = input(f"? {val1} {val2}")
        q-=1
        if val == "<":
            L.append(val1)
            i += 1
        else:
            L.append(val2)
            j += 1
    L += L1[i:] + L2[j:]
    return L, q
      
      

def sort(L, q):
    if L == []:
        return [], q
    if len(L) == 1:
        return L, q
    n = len(L)
    k = n//2
    L1, q = sort(L[:k], q)
    L2, q = sort(L[k:], q)
    L, q = merge(L1, L2, q)
    return L, q

print("!", " ".join(sort(L=liste, q=Q)[0]))
```

```python
N, Q = input().split()
N = int(N)
Q = int(Q)

representant = [i for i in range(N)]

def add_edge(u, v, representant=representant):
  if representant[u] == u:
    representant[u]=v
  else:
    add_edge(representant[u], v)
    
def same_edge(u, v, representant=representant):
  if representant[u] == u:
    if representant[v] == v:
      return u == v
    else:
      return same_edge(u, representant[v])
  else:
    return same_edge(representant[u], v)
  

for q in range(Q):
  a, u, v = input().split()
  u = int(u)
  v = int(v)
  if a == "0":
    add_edge(u, v)
  else:
    if same_edge(u, v):
      print(1)
    else:
      print(0)
    
```
